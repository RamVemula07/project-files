```
## Remove the duplicate elements in list
## ex:[11,22,11,25,36,36,11,25]
## o/p: [11,22,25,36]
```

```
li=[11,22,11,25,36,36,11,25]
l2=[]
for i in li:
  #print(i,end=" ")
  if i not in l2:
    l2.append(i)
print(l2)    
```

```
## Print the even numbers in ascending order and print the numbers in descinding order
## i/p: [1,2,3,4,5,6,7,8,9]
## o/p: [2,4,6,8,9,7,5,3,1]
```

```
l=[1,2,3,4,5,6,7,8,9]
evenlist=[]
oddlist=[]
for n in l:
  if n%2==0:
    evenlist.append(n)
  else:
    oddlist.append(n)
## evenlist
## oddlist 
evenlist+oddlist[::-1] 
```

```
oddlist.sort(reverse=True)
oddlist
```
